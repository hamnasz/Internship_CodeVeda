```python
import pandas as pd
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.preprocessing import StandardScaler, LabelEncoder
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

# 1. Load the dataset

```python
try:
    df = pd.read_csv('data.csv')
    print("Dataset loaded successfully.")
except FileNotFoundError:
    print("Error: 'data.csv' not found. Please ensure the file is in the correct directory.")
    exit()
```

```python
print("\nFirst 5 rows of the dataset:")
print(df.head())
print("\nDataset Info:")
df.info()
print("\nValue counts for 'species':")
print(df['species'].value_counts())
```

```python
X = df.drop('species', axis=1)
y = df['species']
```

```python
le = LabelEncoder()
y_encoded = le.fit_transform(y)
print(f"\nOriginal species labels: {le.classes_}")
print(f"Encoded species labels (first 5): {y_encoded[:5]}")
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.3, random_state=42, stratify=y_encoded)
```

```python
print(f"\nTraining set shape: {X_train.shape}")
print(f"Testing set shape: {X_test.shape}")
```

# 2. Scale the features

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
print("\nFeatures scaled successfully.")
```

# 3. KNN Model Implementation and Evaluation

```python
k_values = [1, 3, 5, 7, 9, 11, 13, 15]
results = {}
```

```python
print("\n--- Evaluating KNN for different K values ---")
```

```python
for k in k_values:
    print(f"\nTraining KNN with K = {k}")
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train_scaled, y_train)

    y_pred = knn.predict(X_test_scaled)

    accuracy = accuracy_score(y_test, y_pred)
    conf_matrix = confusion_matrix(y_test, y_pred)
    class_report = classification_report(y_test, y_pred, target_names=le.classes_, output_dict=True)

    results[k] = {
        'accuracy': accuracy,
        'confusion_matrix': conf_matrix,
        'classification_report': class_report
    }

    print(f"Accuracy for K={k}: {accuracy:.4f}")
    print(f"Confusion Matrix for K={k}:\n{conf_matrix}")
    print(f"Classification Report for K={k}:\n")
    print(pd.DataFrame(class_report).transpose())

    plt.figure(figsize=(6, 5))
    sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues',
                xticklabels=le.classes_, yticklabels=le.classes_)
    plt.title(f'Confusion Matrix for K = {k}')
    plt.xlabel('Predicted Label')
    plt.ylabel('True Label')
    plt.show()
```

# 4. Compare Results

```python
print("\n--- Summary of Results for different K values ---")
accuracy_scores = {k: res['accuracy'] for k, res in results.items()}
print("\nAccuracy Scores:")
for k, acc in accuracy_scores.items():
    print(f"K={k}: {acc:.4f}")
```

# 5. Find the best K based on accuracy

```python
best_k = max(accuracy_scores, key=accuracy_scores.get)
print(f"\nBest K based on accuracy: {best_k} (Accuracy: {accuracy_scores[best_k]:.4f})")
```

```python
print("\nDetailed Classification Report for Best K:")
best_report_df = pd.DataFrame(results[best_k]['classification_report']).transpose()
print(best_report_df)
```

```python
print("\nAnalysis:")
print("The Iris dataset is well-suited for KNN. We observed how accuracy, precision, and recall change with different K values.")
print("Typically, a moderate K value (like 3, 5, or 7) performs well, balancing bias and variance.")
print("Very small K (e.g., K=1) can be sensitive to noise, while very large K can oversimplify the decision boundary.")
```
